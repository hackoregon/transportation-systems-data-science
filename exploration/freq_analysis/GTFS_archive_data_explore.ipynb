# SUMMARY
Okay, let's see what we have that we like.
* calendar_dates.txt
    * Seems to say which route codes were used on which days for this time period. This will tell us which trips to use I think.
* shapes.txt
    * Seems to be shape information for the routes. Likely helpful in plotting the routes.
* stop_times.txt
    * Probably the most helpful. For a given trip, says when the stops are scheduled. This will be where we get frequency information.
* stops.txt
    * Stop information. Likely helpful in plotting stops themselves, and tying to ridership numbers.
* trips.txt
    * Trip information that will tie together the scheduled info, trips, and dates.
    
As of now, I would think the general flow to get frequency info would be as follows.
1. Pick a date of service from calendar_dates.txt
2. For each of the service codes on that date, collect the trip ids used (across all routes).
3. For each trip id, calculate the service frequency based on the number of scheduled stops.

Seems easy! ... I have concerns.
1. I'm not quite sure what the, we'll call them "secondary", stop codes are. I get the S, U, W, but not sure of the others. Hopefully the other letters refer to a small number of special routes that don't overlap with the main ones.
    * Edit: The ridership data only uses code S, U and W. So we'll assume it's okay to just go with that.
2. How to count stop frequency? I guess for now, maybe bin them as the number of stops per hour. We can aggregate for different parts of the day from there, but hourly seems to be a good minimum bin.

Hmm. But wait. This would get us ... frequency by line? That's not what we want. Let's go back to what we're looking to do. We have ridership data. Which is the number of ons / offs at each stop (by line). So, logically, we want the number of buses that stop at each stop (by line)! Let's see what that looks like.
1. Pick a date of service from calendar_dates.txt
2. For each of the service codes on that date, collect the trip ids used (across all routes).
3. For each trip id on each date, for each stop within each trip id, bin the scheduled stop data (by hour).
